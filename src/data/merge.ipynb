```python
import pandas as pd
```

```python
headers = ['Wikipedia movie ID', 'Freebase movie ID', 'Movie name', 'Movie release date', 'Movie box office revenue', 'Movie runtime', 'Movie languages', 'Movie countries', 'Movie genres']

df_movie_metadata = pd.read_csv('../../data/MovieSummaries/movie.metadata.tsv', sep='\t', names=headers, header=None)
print(len(df_movie_metadata))
df_movie_metadata.head()
```

```python
df_movies = pd.read_csv('../../data/ml-32m/movies.csv', sep=',', header=0)
print(len(df_movies))
df_movies.head()
```

```python
df_movies['merge_title'] = df_movies['title'].str.extract(r'^(.*?)(?=\s\()')[0].str.lower()
df_movies['merge_year'] = df_movies['title'].str.extract(r'\((\d{4})\)')[0].astype('Int64')

df_movies.head()
```

```python
df_movie_metadata['merge_title'] = df_movie_metadata['Movie name'].str.lower().str.strip()
df_movie_metadata['merge_year'] = df_movie_metadata['Movie release date'].apply(lambda x: x.split('-')[0] if pd.notna(x) else None).astype('Int64')
df_movie_metadata.head(10)
```

```python
df_merged = pd.merge(df_movie_metadata, df_movies, how='inner', on=['merge_title', 'merge_year'])
print(len(df_merged))
df_merged.head()
```

```python
prefix_path = '../../data/MovieSummaries'
header = ['Wikipedia movie ID', 'Summary']
df_plots = pd.read_csv(f'{prefix_path}/plot_summaries.txt', names=header, sep='\t')
print(len(df_plots))
df_plots.head()
```

```python
df_complete = pd.merge(df_merged, df_plots, how='inner', on=['Wikipedia movie ID'])
print(len(df_complete))
df_complete.head()
```

```python
import ast

def conv_to_dict(val):
    try:
        return ast.literal_eval(val) if pd.notna(val) else {}
    except (ValueError, SyntaxError):
        return {}


def get_list(x):
    return list(x.values())
```

```python
df_complete['Languages'] = df_complete['Movie languages'].apply(conv_to_dict).apply(get_list)
df_complete.head()
```
