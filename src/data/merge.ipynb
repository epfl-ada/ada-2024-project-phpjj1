```python
import pandas as pd
```

```python
prefix_path = '../../data/MovieSummaries'
file_path = f'{prefix_path}/movie.metadata.tsv'
headers = ['Wikipedia movie ID', 'Freebase movie ID', 'Movie name', 'Movie release date', 'Movie box office revenue', 'Movie runtime', 'Movie languages', 'Movie countries', 'Movie genres']
df_movie = pd.read_csv(file_path, sep='\t', names=headers, header=None)
print("Number of datapoints: ", len(df_movie))
df_movie.head()
```

```python
df_movie_ratings = pd.read_csv('../../data/movies_metadata.csv', sep=',', header=0)
print("Number of datapoints: ", len(df_movie_ratings))
df_movie_ratings.head()
```

```python
import re

def prepare_dataset(df, date_col='Movie release date', name_col='Movie name'):
    # Step 1: Create 'merge_year' by extracting the year from the date column
    df['merge_year'] = pd.to_datetime(df[date_col], errors='coerce').dt.year

    # Step 2: Create 'merge_name' by standardizing the movie name
    df['merge_name'] = df[name_col].str.lower()                           # Convert to lowercase
    df['merge_name'] = df['merge_name'].str.strip()                       # Strip whitespace
    df['merge_name'] = df['merge_name'].apply(lambda x: re.sub(r'[^a-z0-9\s]', '', x))  # Remove special characters

    return df
```

```python
df_movie = prepare_dataset(df_movie)
df_movie_ratings = prepare_dataset(df_movie_ratings, 'release_date', 'original_title')
print(f"df movie len: {len(df_movie)}, df ratings len: {len(df_movie_ratings)}")
```

```python
df_merged = pd.merge(df_movie, df_movie_ratings, on=['merge_name', 'merge_year'], how='inner')
print("Number of merged datapoints: ", len(df_merged))
df_merged.head()
```
