```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

# Load THE MOVIE dataset
themovies_ratings = pd.read_csv("../data/THEMOVIE_ds/ratings.csv")
themovies_metadata = pd.read_csv("../data/THEMOVIE_ds/themovies_metadata.csv")
```

```python
print("shape of ratings dataset: ", themovies_ratings.shape)
print("shape of metadata dataset: ", themovies_metadata.shape)
```

```python
#Ids that are bad formatted 
bad_ids = themovies_metadata[~themovies_metadata['id'].apply(lambda x: str(x).isdigit())]

themovies_metadata = themovies_metadata[~themovies_metadata['id'].isin(bad_ids['id'])]
themovies_metadata['id'] = themovies_metadata['id'].astype(int)
```

```python
# Avereage rating of each movie 

summary_ratings = themovies_ratings.groupby('movieId')['rating'].agg(
    avg_rating='mean', 
    num_ratings='count',
    std_rating='std'
).reset_index()
summary_ratings.shape
```

```python

themovie_merged = themovies_metadata.merge(summary_ratings, left_on= 'id', right_on= 'movieId', how = 'left').drop('movieId', axis=1)
themovie_merged.head()
```
